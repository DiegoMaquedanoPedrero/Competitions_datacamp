## 2. Import pre loaded functions, hand made functions and define global variables

```python
def custom_description(df):
    """
    Get distinct values and number of them in a dataframe
    :param df: Dataframe to check
    :return: Df with list and count of unique values
    """
    values = df.apply(lambda col: col.unique())
    counts = df.apply(lambda col: col.nunique())
    nas = df.apply(lambda col: (sum(col.isna()) * 100 / len(df)))
    types = df.apply(lambda col: col.dtype)
    resumen = pd.concat([types, values, counts, nas], axis=1).rename(columns={0: "column_type",
                                                                               1: "unique_values",
                                                                               2: "number_of_unique_values",
                                                                               3: "prop_of_nas"})    
    return (resumen)

#######################################################################################################################################
def remove_rows_with_na(df, proportion):
    """
    Function to remove rows with more than a specified proportion of Nans
    :param df:
    :param proportion: 
    :return: df without those rows and df of rows removed
    """
    df_copy = df.copy()
    rows_to_rm = (df.isnull().sum(axis=1) / len(df.columns.tolist())) > proportion
    result = df_copy[~rows_to_rm]
    removed = df_copy[rows_to_rm]
    return result, removed

#######################################################################################################################################
def variables_distribution_plots(df, vars_to_skip):
    """
    Create barplots for categorical  variables
    :param df: Dataframe with variables to plot
    :param cols_to_skip: list of columns that shouldnt be plotted
    :return: Nothing. Plots 
    """
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    for c in list(df.columns):
        if (df[c].dtype in numerics) and (c not in vars_to_skip):
            sns.set()
            fig, axs = plt.subplots(ncols=2, figsize= (10,5))
            fig.suptitle('Variable distribution')
            sns.ecdfplot(data=df, x=c, ax = axs[0])
            sns.boxplot(x=c, data=df, ax = axs[1])
        
#######################################################################################################################################   
def cap_outliers_percentiles(df,vars_to_skip, perc1=10, perc2=90):
    """
    Replace outliers beyond a specified percentile with the value of that percentile
    """
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    df_result = df.copy()
    
    for c in list(df_result.columns):
        if (df_result[c].dtype in numerics) and (c not in vars_to_skip):
            # Computing 10th, 90th percentiles and replacing the outliers
            floor_percentile = np.nanpercentile(df_result[c], perc1)
            cap_percentile = np.nanpercentile(df_result[c], perc2)
            # print(tenth_percentile, ninetieth_percentile)b = np.where(sample<tenth_percentile, tenth_percentile, sample)
            df_result[c] = np.where(df_result[c] > cap_percentile, cap_percentile, df_result[c])
            df_result[c] = np.where(df_result[c] < floor_percentile, floor_percentile, df_result[c])
    
    return df_result

#######################################################################################################################################
def cap_outliers_percentiles_iqr(df,vars_to_skip):
    """
    Replace outliers with the value of +-1.5 * iqr and shows the number of outliers per variable
    """
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    df_result = df.copy()
    
    for c in list(df_result.columns):
        if (df_result[c].dtype in numerics) and (c not in vars_to_skip):
            # Computing 10th, 90th percentiles and replacing the outliers
            floor_percentile = np.nanpercentile(df_result[c], 25)
            cap_percentile = np.nanpercentile(df_result[c], 75)
            iqr = cap_percentile - floor_percentile
            outliers = (sum(np.where((df_result[c] > cap_percentile + 1.5 * iqr) | (df_result[c] < floor_percentile - 1.5 * iqr),
                                1, 0)) * 100/len(df_result)).round(2)
            # print(tenth_percentile, ninetieth_percentile)b = np.where(sample<tenth_percentile, tenth_percentile, sample)
            df_result[c] = np.where(df_result[c] > cap_percentile + 1.5 * iqr , cap_percentile, df_result[c])
            df_result[c] = np.where(df_result[c] < floor_percentile - 1.5 * iqr, floor_percentile, df_result[c])
            print(f"{c}: {outliers} outliers")
    
    return df_result

#######################################################################################################################################
def remove_non_informative_columns(df, vars_to_skip):
    
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    df_result = df.copy()
    
    for c in list(df_result.columns):
        if (df_result[c].dtype in numerics) and (c not in vars_to_skip):
            zero_prop = pd.DataFrame(df_result[c].value_counts(normalize = True).nlargest(1)).iloc[0,0]
            if zero_prop > 0.95:
                df_result.drop([c], axis = 1, inplace=True)
    
    return df_result

#######################################################################################################################################
def cleaning_whitespaces(df, *args):
    """
    Function to clean trailing and ceiling whitespaces from non numeric columns of a dataframe and other characters
    :param df: 
    :*args: characters to remove from the dataframe columns
    :return: df without the specified characters or whitespaces
    """
    result = df.copy()
    for x in args:
        result = result.apply(lambda col: col.str.replace(x, "").str.strip() if col.dtype not in numerics else col)
    return result
```

Variables which are integers like DAYS_CREDIT_UPDATE should not be modified because:

#### Do we need to convert integer variables to categorical?

"It really depends on context.

If the integer variable has some inherent ordering to it, for example it could be colours where lower numbers represent "darker shades" and higher numbers represented "lighter shades", then treating it as a continuous variable is almost certainly preferable. Not only would it make more sense, but you're eliminating some 200 variables from your model which is a huge bonus.

On the other hand if these integers have no inherent ordering, say for example they represent plots of land, then they should be treated as a categorical variable. It would make no sense to treat them as a continuous variable since its value is independent of the property of the variable you're interested in."

In this case, integers represent ordinal values, it has temporal or numerical meaning so they will be considered as numericals.

### Check for unbalanced categories

```python
for column in bureau_categorical:
    temp = (bureau_categorical[column].value_counts(normalize = True) * 100).reset_index()
    sns.barplot(data = temp, x = "index", y = column)
    plt.xticks(rotation=45)
    plt.show()
```

We can observe that classes are absolutely imbalanced and as the minor categories are absolutely irrelevant, we will keep only 2 categories for each variable

```python
# We use where to replace values matching tbe conditions required
bureau_categorical["CREDIT_ACTIVE"] = np.where(bureau_categorical["CREDIT_ACTIVE"] != "Closed", "Active", "Closed")

bureau_categorical["CREDIT_CURRENCY"] = np.where(bureau_categorical["CREDIT_CURRENCY"] != "currency 1", "other", "currency 1")

bureau_categorical["CREDIT_TYPE"] = np.where((bureau_categorical["CREDIT_TYPE"] != "Consumer credit") & (bureau_categorical["CREDIT_TYPE"] != "Credit card"),
                                      "Other", bureau_categorical["CREDIT_TYPE"] )

bureau_categorical["times_prolongued_credit_bureau"] = np.where(bureau_categorical["times_prolongued_credit_bureau"] != 0, 1, 0)
```

In the next step, we have to update the original dataframe with the resulting values from the previous calculations

```python
for column in bureau_categorical.columns:
    if column in bureau:
        bureau[column] = bureau_categorical[column] 
```
